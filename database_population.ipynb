```python
import sqlite3
conn = sqlite3.connect('sanfrancisco.db')
import datetime
```

```python
#change NOT NULL to NULL create function like below to each

c = conn.cursor()
c.execute('''CREATE TABLE IF NOT EXISTS 
"Address" (
    "Address" string  NOT NULL ,
    "City" string  NOT NULL ,
    "State" string  NOT NULL ,
    "Zipcode" int  NOT NULL ,
    "Location" string  NOT NULL ,
    "Original_Price" int  NULL ,
    "Listing_date" int NULL ,
    "Days_on_Market" int NULL ,
    "Sold_Price" int NULL ,
    "Sold_date" int  NOT NULL ,
    "Bedroom" float  NOT NULL ,
    "Bathroom" float  NOT NULL ,
    "Sq_ft" int  NULL ,
    "Latitude" float  NOT NULL ,
    "Longitude" float  NOT NULL ,
    "URL" string  NOT NULL ,
    CONSTRAINT "pk_Address-PK" PRIMARY KEY (
        "Address"
    )
)''')

conn.commit()
```
